# ***Importing Libraries***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
import glob as gb
from tqdm import tqdm
import cv2 # to read images
```

# ***Get The Images***

```python
#get all image files and sort them
img_files=sorted(os.listdir(img_path))
mask_files=sorted(os.listdir(mask_path))
```

```python
X=[] # to store features
y=[] # to store target
```

# ***Data Preprocessing***

```python
for i in tqdm(range(len(img_files))):
    img_dir=os.path.join(img_path,img_files[i]) # get each path
    mask_dir=os.path.join(mask_path,mask_files[i])

    img=cv2.imread(img_dir) # read the image in this path
    img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert colors from BGR to RGB
    img=cv2.resize(img,(128,128))
    img=img/255.0

    mask = cv2.imread(mask_dir, cv2.IMREAD_GRAYSCALE)
    mask=cv2.resize(mask,(128,128))
    mask=mask/255.0
    mask = np.expand_dims(mask, axis=-1)

    mask = (mask > 0).astype(np.float32)

    X.append(img)
    y.append(mask)
```

```python
X = np.array(X)
y = np.array(y)

print(X.shape)
print(y.shape)
```

# ***Spliting Data***

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(
    X, y, random_state=42,test_size=0.20
)
```

# ***UNet Model***

```python
from tensorflow.keras import layers, models

def simple_unet_model(input_shape=(128, 128, 3)):
    inputs = layers.Input(input_shape) # Define the input layer with specified image shape

    # --- Encoder (Downsampling) ---
    # First block: extract basic features
    c1 = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(inputs) # Apply 1st convolution with 16 filters
    c1 = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(c1)     # Apply 2nd convolution to refine features
    p1 = layers.MaxPooling2D((2, 2))(c1)                                     # Reduce spatial dimensions by half

    # Second block: extract more complex features
    c2 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(p1) # Apply convolution with 32 filters
    c2 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(c2) # Refine features at this scale
    p2 = layers.MaxPooling2D((2, 2))(c2)                                     # Further reduce spatial dimensions

    # --- Bridge (Bottleneck) ---
    # The deepest part of the network
    c3 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(p2) # Deepest feature extraction with 64 filters
    c3 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c3) # Refine the bottleneck features

    # --- Decoder (Upsampling) ---
    # First upsampling block: reconstruct spatial resolution
    u4 = layers.Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c3) # Upscale features back to previous size
    u4 = layers.concatenate([u4, c2])                                           # Skip connection: combine encoder features
    c4 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(u4)       # Process combined features
    c4 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(c4)       # Refine reconstructed features

    # Second upsampling block: return to original size
    u5 = layers.Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same')(c4) # Upscale back to original input resolution
    u5 = layers.concatenate([u5, c1])                                           # Skip connection: combine early encoder features
    c5 = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(u5)       # Process features at original resolution
    c5 = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(c5)       # Final refinement of features

    # --- Output Layer ---
    # Generate the final segmentation mask
    outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(c5) # Final 1x1 conv to produce binary mask (0 or 1)

    model = models.Model(inputs=[inputs], outputs=[outputs]) # Create the Keras model instance
    return model
```

```python
# Create the model
model = simple_unet_model() # Instantiate the simple UNet model
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) # Setup training configuration
model.summary() # Print the network architecture summary
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=8, epochs=10)
```

# ***Visualization The Result***
